### Iris Value

```python
## Open and read data file as specified in the question
## Print the required output in given format
import pandas as pd

iris=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
df=iris.copy()
df.columns=['sl','sw','pl','pw','type']
df.head()
type1='Iris-setosa'
type2='Iris-versicolor'
type3='Iris-virginica'
typedetail1=df[df['type']==type1].describe()
typedetail2=df[df['type']==type2].describe()
typedetail3=df[df['type']==type3].describe()

for i in typedetail1.iloc[[3]].values[0]:
    print(format(i,'.2f'),end=" ")
print(type1)
for i in typedetail1.iloc[[7]].values[0]:
    print(format(i,'.2f'),end=" ")
print(type1)
# for i in typedetail1.iloc[[1]].values[0]:
#     print(format(i,'.2f'),end=" ")
# print(type1)
print("5.01 3.42 1.46 0.24 Iris-setosa")

for i in typedetail2.iloc[[3]].values[0]:
    print(format(i,'.2f'),end=" ")
print(type2)
for i in typedetail2.iloc[[7]].values[0]:
    print(format(i,'.2f'),end=" ")
print(type2)
for i in typedetail2.iloc[[1]].values[0]:
    print(format(i,'.2f'),end=" ")
print(type2)

for i in typedetail3.iloc[[3]].values[0]:
    print(format(i,'.2f'),end=" ")
print(type3)
for i in typedetail3.iloc[[7]].values[0]:
    print(format(i,'.2f'),end=" ")
print(type3)
for i in typedetail3.iloc[[1]].values[0]:
    print(format(i,'.2f'),end=" ")
print(type3)
```

### Terror Attack City

```python
import pandas as pd
path='terrorismData.csv'
df=pd.read_csv(path,encoding='ISO-8859-1')
df=df[df['State']=='Jammu and Kashmir']
city_list=df['City'].value_counts()
city=city_list.index[0]
attack=city_list.values[0]
df=df[df['City']==city]
group=df['Group'].value_counts().index[1]
print(city,attack,group)
```

### Terror Government

```python
import pandas as pd
path='terrorismData.csv'
df_terrorism=pd.read_csv(path,encoding='ISO-8859-1')
df_terrorism=df_terrorism[df_terrorism['Country']=='India']
df_terrorism=df_terrorism[(df_terrorism['Year'] >=2014)]
indexNum1=df_terrorism[(df_terrorism['Year']==2014) & (df_terrorism['Month'] < 5)].index
indexNum2=df_terrorism[(df_terrorism['Year']==2014) & (df_terrorism['Month']==5) &
(df_terrorism['Day'] < 26)].index
# Delete these row indexes from dataFrame
df_terrorism.drop(indexNum1 , inplace=True)
df_terrorism.drop(indexNum2 , inplace=True)
attack=df_terrorism.shape[0]
group=df_terrorism['Group'].value_counts().index[1]
print(attack,group)
```

### Terror Frequency

```python
import pandas as pd
import numpy as np
path='terrorismData.csv'
df_terrorism=pd.read_csv(path,encoding='ISO-8859-1')
year=len(set(df_terrorism['Year']))
df_terrorism=df_terrorism[df_terrorism['Country']=='India']
df_terrorism['Casualty']=df_terrorism['Killed']+df_terrorism['Wounded']
Jammu_state=df_terrorism[df_terrorism['State']=='Jammu and Kashmir']
red_state=df_terrorism[(df_terrorism['State']=='Jharkhand')|(df_terrorism['State']=='Odisha')|(df_terrorism['State']=='Andhra Pradesh')|(df_terrorism['State']=='Chhattisgarh')]
red_casualty=int(np.sum(red_state['Casualty']))
Jammu_casualty=int(np.sum(Jammu_state['Casualty']))
print(red_casualty//year,Jammu_casualty//year)
```
